The data can be uniquely identified by grouping them by country and year.  Any duplicates occur if the country and year is the same for any entry.  In filling out data, it may not be appropriate to just do a general average.  We can fill empty values with the country's average over the years. This is a set of time series data.  To match changing trends in the data (growth over the years, decline over the years, stability over the years etc), we can fill missing data with the rolling on 5 years by country.

```python
missing = df.columns[df.isna().any()]
```

```python
missing
```

```python
for item in missing:
    df[item]=df[item].fillna(df.groupby("Country")[item].apply(lambda x:x.rolling(6, min_periods=1).mean()))
```

```python
df.describe()
```

Not all data is replaced using that method.  This may be because of missing data covering more than the five year interval.  Let us fill the remaining based on the general mean based on the country.

```python
for item in missing:
    df[item]=df[item].fillna(df.groupby("Country")[item].transform(lambda x: x.mean()))
df.describe()
```

```python
df.head()
```

Still some data is missing.  Some countries lack data for particular fields.

```python
missing=df.columns[df.isnull().any()]
df[df.isnull().any(axis=1)][missing.insert(0,'Country')]
```

```python
df.loc[df['Country'] == "Australia"].describe()
```

```python
df.loc[df['Country'] == "United Kingdom"].describe()
```

We can treat those fields as if these countries do not produce those particular emmissions.  Alternatively, there may just not be data about those industries in those countries.  Either way, we can handle that missing data by either filling the missing values with zeros or filling it  with the mean value of all the values in the field.  Let us replace missing values with the mean of all countries based on year

```python
for item in missing:
    df[item]=df[item].fillna(df.groupby("Year")[item].transform(lambda x: x.mean()))
```

```python
df.drop_duplicates(keep=False, inplace=True)
```

We also drop any duplicate values if they exist.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

We will use linear regression and decision tree regression to create a model to predict greenhouse gas emmission.

```python
df2= df.iloc[:,[0,1,2,8,9,10,11,12,13]]
encoder = LabelEncoder()
df2["Country"] = encoder.fit_transform(df2["Country"])
df2["Year"] = encoder.fit_transform(df2["Year"])
df2.head()
```

```python
x=df2.drop("Total GHG Emissions Excluding LUCF (MtCO2e)", axis=1).copy().values
y=df2["Total GHG Emissions Excluding LUCF (MtCO2e)"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
decision_tree = DecisionTreeRegressor()
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_test)
```

```python
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
```

```python
reg = LinearRegression().fit(x_train, y_train)
y_pred=reg.predict(x_test)
```

```python
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
```

We used RMSE and MSE to evaluate performance of the models.  We can see that linear regression is the better model as it produces the smaller error results.
